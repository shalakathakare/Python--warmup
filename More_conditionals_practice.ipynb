 Coffee Break: You’re deciding whether you can step out for a coffee. On weekends you always can. On weekdays you can only go if there isn’t a meeting scheduled right now.

```python
def lg_coffee(day, meeting):    

    weekend = ['Saturday', 'Sunday']

    if day in weekend:
        return True
    else:
        if meeting == 1:
            return False
    return True
```

Meeting Clash: You are given two meetings, each with a start time and an end time measured in minutes from midnight. Determine whether the meetings actually overlap in time. If one ends exactly when the other begins, that does not count as overlapping.

```python
def overlap(s1,e1,s2,e2):    
    if (s1 < s2 < e1) or (s2 < s1 < e1):
        return True
    return False
    
```

```python
overlap(10,11,9,10.5)
```

Triangle Validity and Type: You receive three positive numbers that claim to be the lengths of a triangle’s sides. First decide whether they can form a valid triangle at all. If they can, classify the triangle as equilateral (all sides equal), isosceles (exactly two sides equal), or scalene (all sides different). If they cannot form a triangle, say so.

```python
def triangle(side1,side2,side3):
    
    if ((side1+side2)>side3) and ((side1+side3)>side2) and ((side3+side2)>side1):
        if side1 == side2 == side3:
            print("Equilateral triangle")
        elif (side1-side2-side3 == -side3) or (side1-side2-side3 == -side2) or (side2-side3-side1 == -side1):
            print("Isosceles triangle")
        elif (side1!=side2!=side3):
            print("Scalene triangle")
            
    else:
        print("not a valid triangle")
```

```python
triangle(4,10,4)
```

Time-of-Day Label: Given an hour of the day on a 24-hour clock (an integer between 0 and 23), label it as “night,” “morning,” “afternoon,” or “evening.” Hours 22–23 and 0–4 are “night,” 5–11 are “morning,” 12–16 are “afternoon,” and 17–21 are “evening.”

```python
def timeofday(hr):
    
    if (0<hr<5) or (24>hr>20):
        print("night")
    elif (4<hr<12):
        print("morning")
    elif (12<hr<16):
        print("afternoon")
    else:
        print("evening")
    
```
