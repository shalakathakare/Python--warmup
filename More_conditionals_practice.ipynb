6. Meeting Clash: You are given two meetings, each with a start time and an end time measured in minutes from midnight. Determine whether the meetings actually overlap in time. If one ends exactly when the other begins, that does not count as overlapping.

7. Triangle Validity and Type: You receive three positive numbers that claim to be the lengths of a triangle’s sides. First decide whether they can form a valid triangle at all. If they can, classify the triangle as equilateral (all sides equal), isosceles (exactly two sides equal), or scalene (all sides different). If they cannot form a triangle, say so.

```python
def triangle(side1,side2,side3):
    
    if ((side1+side2)>side3) and ((side1+side3)>side2) and ((side3+side2)>side1):
        if side1 == side2 == side3:
            return "Equilateral triangle"
        elif (side1-side2-side3 == -side3) or (side1-side2-side3 == -side2) or (side2-side3-side1 == -side1):
            return "Isosceles triangle"
        elif (side1!=side2!=side3):
            return "Scalene triangle"
            
    return False
```

```python
assert triangle(4,10,4), "not a valid triangle"
```

8. Time-of-Day Label: Given an hour of the day on a 24-hour clock (an integer between 0 and 23), label it as “night,” “morning,” “afternoon,” or “evening.” Hours 22–23 and 0–4 are “night,” 5–11 are “morning,” 12–16 are “afternoon,” and 17–21 are “evening.”

```python
def timeofday(hr):
    
    if (0<hr<5) or (24>hr>20):
        return "night"
    elif (4<hr<12):
        return "morning"
    elif (12<hr<16):
        return "afternoon"
    else:
        return "evening"
    
```

9. Interval Relationship: You are given two closed numeric intervals, each described by two endpoints that might be in any order. After interpreting each interval correctly (left endpoint ≤ right endpoint), report whether the first is entirely inside the second, the second is entirely inside the first, the two merely overlap without containment, or they are disjoint.

```python
def intervals(p1, p2, q1, q2):
    
    if p2 <= p1:
        p1, p2 = p2, p1
    if q2 <= q1:
        q1, q2 = q2, q1
        
    if (p1 < q1 < p2) and (p1 < q2 < p2):
        return "second entirely inside first"
    elif (q1 < p1 < q2) and (q1 < p2 < q2):
        return "first entirely inside second"
    elif (p1 < q1 < p2) or (p1 < q2 < p2):
        return "overlap without containment"
    else:
        return "disjoint"
```

```python
intervals(10,2,4,7)
```

12. Nature of a Triangle by Angle: Given three positive side lengths that do form a valid triangle, decide whether the triangle is right-angled, acute, or obtuse. Base your decision on a comparison involving the square of the longest side and the sum of the squares of the other two sides.

```python
def tri_angle(a,b,c):
    ls = [a,b,c]
    ls.sort(reverse = True)
    d,e,f = ls[0],ls[1],ls[2]
    
    if (d**2 == e**2 + f**2):
        triangle = "right-angled triangle"
    elif (d**2 > e**2 + f**2):
        triangle = "obtuse triangle"
    else:
        triangle = "acute triangle"
        
    return triangle
    
    
```

```python
tri_angle(4,5,3)
```

14. Report on Quadratic Roots: You are given coefficients a, b, and c (with a nonzero) for a quadratic equation. If there are real roots, report them in ascending order; if the roots are complex, simply state that they are complex without computing them.



```python
def quadratic(a,b,c):
    
    d =  (b**2) - (4*a*c)
    
    if d > 0:
        return "two distict real roots"
    elif d == 0:
        return "one repeated real root"
    else:
        return "complex (non-real) roots"
        
```

The format must be exact, including a single colon and two digits on each side; for example, "07:05" is valid while "7:05" is not.
