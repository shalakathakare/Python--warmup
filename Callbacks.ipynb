Write a function that calculates the summation of a series based on a formula.

 sigma(start_n, end_n, term_func)




```python
def num(n):
    return n
```

```python
def num_sq(n):
    return n*n
```

```python
def num_harmonic(n):
    return (1/n)
```

```python
def sigma(start_n, end_n, term_func):

    summ = 0
    for i in range(start_n,end_n + 1 +1 ):       #inclusive of end_n
        op = term_func(i)                        #function is called as per parameter that is passed
        summ = summ + op                         

    return summ
        
```

```python
sigma(1,4,num)
```

```python
sigma(1,4,num_sq)
```

```python
sigma(1,4,num_harmonic)
```

Write a function compute_average_loss(predictions, targets, loss_func) that measures how "wrong" a set of predictions is, based on a customizable rule.

Inputs:

predictions: A list of numbers (e.g., [10, 12, 18]).
targets: A list of correct answers (e.g., [10.1, 12.3, 19]).
loss_func: A function that takes two numbers (pred, target) and returns a single number representing the error.


Task:

Create two lists, the first can be a set of random values, the second can be the same with a bit of noise added to it.
Zip the two lists together, pass the pair into loss_func, sum up the results, and divide by the count to get the average.
You can use this to calculate the behaviour of MAE and MSE as the amount of noise (between the prediction and target) changes

```python
def loss_mse(pred, target):

    l = (pred - target) ** 2                                          #calculate squared error for mse
    
    return l
```

```python
def loss_mae(pred, target):

    l = abs(pred - target)                                            #calculate absolute error for mae
    
    return l
```

```python
def compute_average_loss(predictions, targets, loss_func):

    loss = 0
    for p, t in zip(predictions, targets):

        single_loss = loss_func(p,t)                                  #calculate error for each prediction vs. target,calculates mae or mse depending on function callback
        loss = loss + single_loss                                     #calculate sum of all losses(mse or mae)
        
    result = loss/len(targets)                                        #calculate mean for respective loss function
        
    return result

        
```

```python
#Generate sampple data

import random

# Original list
original = [random.randint(10, 50) for i in range(10)]

# Noisy list
noise_level = 3
noisy = [x + random.randint(-noise_level, noise_level) for x in original]

print(original)
print(noisy)
```

```python
compute_average_loss(noisy, original, loss_mse)
```

```python
compute_average_loss(noisy, original, loss_mae)
```

```python
#Generate different sample sizes with adjustable noise level for predictions and calculate mse and mae values

size = [10,100,1000,10000,100000, 1000000,10000000]   
noise_level = 3
mse_val_3 = [] 
mae_val_3 = []

for s in size:

    og_nums = [random.randint(1,100) for i in range(s)]                                        #generate target values for different sample sizws
    noisy_nums = [n + random.uniform(-noise_level, noise_level) for n in og_nums]              #add noise to target values

    mse_val_3.append(round(compute_average_loss(noisy_nums, og_nums, loss_mse),2))               #append mse value for each sample size to list
    mae_val_3.append(round(compute_average_loss(noisy_nums, og_nums, loss_mae),2))               #append mse value for each sample size to list


print(mse_val_3)
print(mae_val_3)

    
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(size, mse_val_3, marker='o', label = "MSE")
plt.plot(size, mae_val_3, marker='o', label = "MAE")
plt.xscale('log')
plt.xlabel("Sample sizes log scale")
plt.ylabel("Loss values")
plt.title("MSE vs MAE for different sample sizes(noise level = 3)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

size = [10,100,1000,10000,100000,1000000,10000000]
noise_levels = [1, 2, 3]

results = {}

for s in size:
    
    
    og_nums = [random.randint(1,100) for _ in range(s)]
    
    for noise_level in noise_levels:
        
        
        if noise_level not in results:
            results[noise_level] = {"mse": [], "mae": []}
        
        noisy_nums = [n + random.uniform(-noise_level, noise_level) for n in og_nums]
        
        mse = round(compute_average_loss(noisy_nums, og_nums, loss_mse), 2)
        mae = round(compute_average_loss(noisy_nums, og_nums, loss_mae), 2)
        
        results[noise_level]["mse"].append(mse)
        results[noise_level]["mae"].append(mae)
```

```python
plt.figure()

for noise in noise_levels:
    plt.plot(size, results[noise]["mse"], marker='o', label=f"MSE (noise={noise})")
    plt.plot(size, results[noise]["mae"], linestyle='--', label=f"MAE (noise={noise})")

plt.xscale('log')
plt.xlabel("Sample sizes (log scale)")
plt.ylabel("Loss values")
plt.title("MSE and MAE Across Noise Levels")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Adjusting noise randomly using the gaussian method 

import random
import matplotlib.pyplot as plt

size = [10,100,1000,10000,100000,1000000,10000000]
noise_levels = [1, 2, 3]

results = {}

for s in size:
    
    
    og_nums = [random.randint(1,100) for _ in range(s)]
    
    for noise_level in noise_levels:
        
        
        if noise_level not in results:
            results[noise_level] = {"mse": [], "mae": []}
        
        noisy_nums = [n + random.gauss(0, noise_level) for n in og_nums]
        
        mse = round(compute_average_loss(noisy_nums, og_nums, loss_mse), 2)
        mae = round(compute_average_loss(noisy_nums, og_nums, loss_mae), 2)
        
        results[noise_level]["mse"].append(mse)
        results[noise_level]["mae"].append(mae)
```

```python
plt.figure()

for noise in noise_levels:
    plt.plot(size, results[noise]["mse"], marker='o', label=f"MSE (noise={noise})")
    plt.plot(size, results[noise]["mae"], linestyle='--', label=f"MAE (noise={noise})")

plt.xscale('log')
plt.xlabel("Sample sizes (log scale)")
plt.ylabel("Loss values")
plt.title("MSE and MAE Across Noise Levels")
plt.grid(True)
plt.legend()
plt.show()
```

Inference: 

MSE values are consistently greater than MAE values for all noise levels and sizes of data points.
