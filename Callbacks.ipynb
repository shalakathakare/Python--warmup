Write a function that calculates the summation of a series based on a formula.

 sigma(start_n, end_n, term_func)




```python
def num(n):
    return n
```

```python
def num_sq(n):
    return n*n
```

```python
def num_harmonic(n):
    return (1/n)
```

```python
def sigma(start_n, end_n, term_func):

    summ = 0
    for i in range(start_n,end_n + 1):
        op = term_func(i)
        summ = summ + op

    return summ
        
```

```python
sigma(1,4,num)
```

```python
sigma(1,4,num_sq)
```

```python
sigma(1,4,num_harmonic)
```

Write a function compute_average_loss(predictions, targets, loss_func) that measures how "wrong" a set of predictions is, based on a customizable rule.

Inputs:

predictions: A list of numbers (e.g., [10, 12, 18]).
targets: A list of correct answers (e.g., [10.1, 12.3, 19]).
loss_func: A function that takes two numbers (pred, target) and returns a single number representing the error.


Task:

Create two lists, the first can be a set of random values, the second can be the same with a bit of noise added to it.
Zip the two lists together, pass the pair into loss_func, sum up the results, and divide by the count to get the average.
You can use this to calculate the behaviour of MAE and MSE as the amount of noise (between the prediction and target) changes

```python
def loss_mse(pred, target):

    l = (pred - target) ** 2
    
    return l
```

```python
def loss_mae(pred, target):

    l = abs(pred - target)
    
    return l
```

```python
def compute_average_loss(predictions, targets, loss_func):

    loss = 0
    for i, j in zip(predictions, targets):

        single_loss = loss_func(i,j)
        loss = loss + single_loss
        
    result = loss/len(targets)
        
    return result

        
```

```python
import random

# Original list
original = [random.randint(10, 50) for i in range(10)]

# Noisy list
noisy = [x + random.randint(-3, 3) for x in original]

print(original)
print(noisy)
```

```python
compute_average_loss(noisy, original, loss_mse)
```

```python
compute_average_loss(noisy, original, loss_mae)
```
