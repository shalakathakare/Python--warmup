```python
print("Enter a list of integers seperated by space:")

lst = list(map(int, input().split()))
```

### 2. Write a function that takes a list of integers and calculates the sum of all the odd values. Return the sum and the number of values encountered.

```python
print("Enter a list of integers seperated by space:")

lst = list(map(int, input().split()))
```

```python
def sum_odd(lint):   
    
    sum = 0
    for i in lint:
        if i%2!=0:
            sum = sum + i
       
    return sum
```

### 5. Implement factorial iteratively and recursively.

```python
def factorial():
    n = int(input("Enter a number to calculate factorial:"))
    fact = 1
    for i in range(n,1,-1):
        fact = fact * i
    print(fact)
```

### 6. Calculate the nth Fibonacci number iteratively and recursively.

```python
def fibonacci():
    n = int(input("Enter a number to calculate fibonacci:"))
    seq = [0,1]
    for i in range(1,n+1):
        seq.append(seq[i]+seq[i-1])
    return seq[n-1]
        
```

```python
fibonacci()
```

### 7. Write a function that checks if a given word is a palindrome.

```python
def check_pal():   
    word = str(input("Enter a word to check for palindrome:"))
    lw = len(word)
    mp = (len(word))//2
    lst = []
    for i in range(mp+1):
        for k in range(lw-1,mp,-1):
            if word[i] == word[k]:
                lst.append(True)
            else:
                lst.append(True)
    if all(lst):
        print(word,"is a palindrome")
    else:
        print(word,"is not a palindrome")
    
```

```python
check_pal()
```

### 10. Check if a Number is an Armstrong Number. An Armstrong number is a number that is the sum of its own digits each raised to the power of the number of digits. For example, 153 = 1^3 + 5^3 + 3^3.

```python
def check_armstrong():
  n = int(input("Enter number to check for Armstrong number: "))
  pow = len(str(n))
  sum = 0
  for i in str(n):
    ind = int(i)**pow
    sum = sum+ind

  if sum == n:
    print(n,"is an Armstrong number")
  else:
    print(n,"is not an Armstrong number")
```

```python
check_armstrong()
```

### 11. Take a 3 digit number from the user, with different first and last digits. Reverse it and subtract the smaller from the larger. Call this d. Reverse d and add it to d. Print the result

```python
#Long route

n = str(input("Enter a 3 digit number with different first and last digits "))
n1 = int(n)
n2 = list(n)
n2.reverse()
n2 = int(''.join(n2))

if n1>n2:
    d = n1-n2
else:
    d = n2-n1
    
d1 = list(str(d))
d1.reverse()
d1 = int(''.join(d1))

res = d+d1

print(res)


```

```python
#Math route

def reverse_num(num):
    ln = len(str(num)) - 1
    lst = []
    while ln >= 0:
        q = num//(10**ln)
        r = num%(10**ln)
        lst.append(q)
        ln = ln-1
        num = r
        
    lst.reverse()
    
    lrn = len(lst) - 1
    
    rl = []
    for i in lst:
        sr = i*(10**lrn)
        rl.append(sr)
        lrn = lrn - 1
              
    return sum(rl)


```

```python
def solve_d(n):  
    rev_num = reverse_num(n)

    if n>rev_num:
        d = n - rev_num
    else:
        d = rev_num - n
    
    
    rev_d = reverse_num(d)

    res = d + rev_d
    
    return res

```

```python
n = int(input("Enter a 3 digit number with different first and last digits "))
solve_d(n)
```
