####  Check if a number is even or odd

```python
def even_odd(num):
    if num%2 == 0:
        return "even"
    else:
        return "odd"
    
```

```python
even_odd(0)
```

#### Check if a number is positive or negative or zero

```python
def pos_neg(num):
    if num == 0:
        return "zero"
    elif num > 0:
        return "positive"
    else:
        return "negative"
    
```

```python
pos_neg(0)
```

#### Print the number of days in a year

```python
def check_leap(n):
    if (n%4 == 0 and n%100 != 0) or  (n%4 == 0 and n%400 == 0):
        return True
    else:
        return False
```

```python
def num_days(yr):
    if check_leap(yr) == True:
        return 366
    else:
        return 365
```

```python
num_days(1976)
```

#### Print the number of days in a month

```python
def days_month(mnth):
    d30 = ["September", "April", "June", "November"]
    d31 = ["Januray", "March", "May", "July", "October", "December"]
    if mnth in d30:
        days = 30
    elif mnth in d31:
        days = 31
    else:
        days =28
    return days
    
```

```python
days_month("February")
```

#### Check number of days in a month according to leap year

```python
def days_month_ly(mnth, yr):
    d30 = ["September", "April", "June", "November"]
    d31 = ["Januray", "March", "May", "July", "October", "December"]
    if mnth in d30:
        days = 30
    elif mnth in d31:
        days = 31
    else:
        if check_leap(yr) == True:
            days = 29
        else:
            days = 28
    return days
```

```python
days_month_ly("February", 1900)
```

#### Convert HH:MM:SS to seconds

```python
def time_sec(h,m,s):
    if m>59 or s>59 :
        return "invalid input"
    
    sec = h * 60**2 + m * 60**1 + s * 60**0
    return sec
```
