Mean and standard deviation

Generate a 1 dimensional array of  integers, say between 10-20. There are many instances when we will want to ‘normalize’ these to a specific range, say between 0-1. This involves calculating the mean and subtracting it from every number.

```python
import numpy as np
```

```python
np.random.seed(42)
arr = np.random.randint(10,20,size = 30)
print(arr)
```

```python
def loop_mean(arr):
    
    summ = 0
    for i in arr:
        summ = summ + i
    l_mean = summ/len(arr)

    return l_mean
    
```

```python
loop_mean(arr)
```

```python
np_mean = np.mean(arr)
print(np_mean)
```

```python
norm_arr = arr - np_mean
print(norm_arr)
```

```python
def loop_std(arr):
    summ_sq = 0
    mean = loop_mean(arr)
    for i in arr:
        err_sq = (i - mean) ** 2
        summ_sq = summ_sq + err_sq

    std = (summ_sq/len(arr)) ** 0.5

    return std
```

```python
loop_std(arr)
```

```python
np_std = np.std(arr)             
print(np_std)
```

```python
loop_std(norm_arr)
```

```python
np_std_norm = np.std(norm_arr)             
print(np_std_norm)
```
