Import the math library. You can use math.sin() and math.cos() to calculate the 
 and 
 of the argument. Create a list of $ sin $ and $ cos $ of numbers from 0.1 … 1.0 and 0.01 … 1.0

```python
import math

lst_sin = [math.sin(i/10) for i in range(1,11)]
lst_cos = [math.cos(i/10) for i in range(1,11)]
```

```python
print(lst_sin)
print(lst_cos)
```

$ tan $ is easily calculated using  $ tan x $ = $ sin x $/$ cos x $
. Create a list of tan() of numbers from 0.1 … 1.0 and 0.01 … 1.0

```python
lst_tan = [math.sin(i/10)/math.cos(i/10) for i in range(1,11)]
```

```python
print(lst_tan)
```
