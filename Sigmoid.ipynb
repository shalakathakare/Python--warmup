```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
## 1-D Sigmoid
x = np.linspace(-10, 10, 100)
a = 1
y = 1 / (1 + a * np.exp(-x))
plt.plot(x, y)
plt.title('1-D Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
a_values = [0.1, 0.5, 1, 2, 5, 10]
x = np.linspace(-10, 10, 100)
for i, a in enumerate(a_values):
    y = 1 / (1 + np.exp(-1 * a * x))
    ax[i//3, i%3].plot(x, y)
    ax[i//3, i%3].axhline(0.5, color='blue', linestyle='--', linewidth=1)  
    ax[i//3, i%3].set_title(f'1-D Sigmoid Function (a={a})')
    ax[i//3, i%3].set_xlabel('x')
    ax[i//3, i%3].set_ylabel('sigmoid(x)')
    ax[i//3, i%3].grid()
plt.tight_layout()
plt.show()
```
